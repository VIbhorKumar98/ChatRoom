```
def applying_rules(file_name,test_df):
  file = open(file_name)
  Lines = file.readlines() 

  predicted_label = []

  rule_values = []
  test_df_rows = len(test_df['id'])
  
  
  for index in range(test_df_rows):
    test_rules = test_df.iloc[index].values.tolist()
    

    for line in Lines:

      rule = line.split('==>')
      if(len(rule)!=2):
        break

      else:
        r_label = rule[1]
        r_rule = rule[0]
      
        conditions = r_rule.split()
        rule_values.clear()
        for i in range(len(conditions)):
          each_col = conditions[i].split('=')
          rule_values.append(each_col[1])


        if(set(rule_values).issubset(set(test_rules))):
          predicted_label.append(int(r_label))
          break
  
  print(predicted_label)
  print(len(predicted_label))
      
  return predicted_label

pr = applying_rules('/content/drive/MyDrive/DMGAssignment/Ass3/clean_4503.txt',test_df)    
print(len(pr))
```
